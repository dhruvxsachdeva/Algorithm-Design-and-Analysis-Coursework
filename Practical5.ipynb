```python
import subprocess
import matplotlib.pyplot as plt
```

```python
n_values = [100, 500, 1000, 2000, 4000]  # bigger n
capacity = 10000  # bigger W
runs = 3  # average over multiple runs

frac_times = []
zero_one_times = []

for n in n_values:
    # Fractional Knapsack
    times = []
    for _ in range(runs):
        result = subprocess.run(['./frac.exe', str(n), str(capacity)],
                                stdout=subprocess.PIPE, text=True)
        times.append(float(result.stdout.strip()))
    frac_times.append(sum(times)/runs)

    # 0-1 Knapsack
    times = []
    for _ in range(runs):
        result = subprocess.run(['./knap01.exe', str(n), str(capacity)],
                                stdout=subprocess.PIPE, text=True)
        times.append(float(result.stdout.strip()))
    zero_one_times.append(sum(times)/runs)
```

```python
plt.plot(n_values, frac_times, label='Fractional Knapsack', marker='o')
plt.plot(n_values, zero_one_times, label='0-1 Knapsack', marker='o')
plt.xlabel('Number of items (n)')
plt.ylabel('Execution time (Âµs)')
plt.title('Knapsack Execution Time Comparison')
plt.legend()
plt.grid(True)
plt.show()
```
