#### Mergesort

```python
import subprocess
import matplotlib.pyplot as plt

sizes_merge= [10,50,500,1000, 5000, 10000, 20000, 50000, 100000]
times_merge = []

for n in sizes_merge:
    # Run compiled C program with n as argument
    result = subprocess.run(["./mergesort.exe", str(n)], capture_output=True, text=True)
    times_merge.append(float(result.stdout.strip()))

plt.plot(sizes_merge, times_merge, marker="o")
plt.xlabel("n (number of elements)")
plt.ylabel("Time (s)")
plt.title("Merge Sort Performance")
plt.grid(True)
plt.show()
```
