```python
import matplotlib.pyplot as plt
```

```python
n_number=[100,200,300,400,500,2000]
time=[0.059000,0.424000,0.723000,2.979000,3.070000,93.033000]
plt.plot(n_number,time)
plt.xlabel("Number of elements in array")
plt.ylabel("Time in seconds")
plt.title("Analysis for recursive matrix multiplication")
plt.grid(True)
plt.show
```
