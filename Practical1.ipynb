### Graph of number of elements vs time for linear search

```python
import matplotlib.pyplot as plt
```

```python
n_number=[100,500,1000,5000,10000,100000]
time=[0,0,0,0.000010,0.000017,0.000017]
plt.plot(n_number,time)
plt.xlabel("Number of elements in array")
plt.ylabel("Time in seconds")
plt.title("Analysis for linear search")
plt.grid(True)
plt.show
```

### Graph of number of elements vs time for binary search

```python
n_number=[100,500,1000,5000,10000]
time=[150,190,180,180,210]
plt.plot(n_number,time)
plt.xlabel("Number of elements in array")
plt.ylabel("Time in nano-seconds")
plt.title("Analysis for linear search")
plt.legend()
plt.grid(True)
plt.show
```
