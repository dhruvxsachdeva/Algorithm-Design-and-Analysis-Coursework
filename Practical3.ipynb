```python
import matplotlib.pyplot as plt
```

```python
recursive_num=[10,20,30,40,50]
recursive_time=[0,0.001000,0.06000,7.362000,97.221000]
iterative_num=[10,20,50,100,500]
iterative_time=[0.001000,0.002000,0.004000,0.006000,0.028000]
memo_num=[100,200,500,1000]
memo_time=[0.006000,0.010000,0.022000,0.039000]
tab_num=[50,100,500,1000]
tab_time=[0.004000,0.006000,0.023000,0.037000]
```

```python
plt.figure(figsize=(10,6))
plt.plot(recursive_num, recursive_time, marker='o', label='Recursive')
plt.plot(iterative_num, iterative_time, marker='s', label='Iterative')
plt.plot(memo_num, memo_time, marker='^', label='Memoization')
plt.plot(tab_num, tab_time, marker='d', label='Tabulation')
plt.yscale('log')
plt.xlabel('n (Fibonacci Number Index)')
plt.ylabel('Time (seconds)')
plt.title('Fibonacci Algorithm Performance Comparison')
plt.legend()
plt.grid(True)

plt.show()
```
